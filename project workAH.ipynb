```python
# Origin airport:
origins = ["HOU-iata","JFK-iata","LAX-iata","DFW-iata","DEN-iata"]

# Destination airports: 
destinations = ["HOU-iata","JFK-iata","LAX-iata","DFW-iata","DEN-iata","ATL-iata","ORD-iata","CLT-iata","LAS-iata","PHX-iata","PCO-iata","SEA-iata","MIA-iata","FLL-iata","SFO-iata","EWR-iata","MSP-iata","DTW-iata","BOS-iata","SLC-iata","PHL-iata","BWI-iata","TPA-iata","SAN-iata","AMA-iata","IAD-iata","BNA-iata","ABQ-iata","PDX-iata","DSM-iata","MSY-iata","BIL-iata","FAR-iata","MAF-iata","LBB-iata","BTV-iata","ROA-iata","MKE-iata","MSN-iata","GRB-iata","CYS-iata","JAC-iata","ELP-iata","CRP-iata","BHM-iata","HSV-iata","TUS-iata","LIT-iata","FAT-iata","SMF-iata","GJT-iata","BDL-iata","EYW-iata","PFN-iata","BOI-iata","EVV-iata","AZO-iata","AUS-iata","SAT-iata","MAF-iata"]

# Enter date as YYYY-MM-DD:
date = "2021-09-03"

# Other parameters
currancy = "USD"
originCountry = "US"
locale = "en-US"
baseURL = "https://skyscanner-skyscanner-flight-search-v1.p.rapidapi.com/apiservices/browseroutes/v1.0/"
```

## Loop through destinations and origins to obtain list of flights

```python
airports = {}
flights = {}
listlist = []
from time import sleep

for destination in destinations:
    for origin in origins:
        
        myurl = baseURL + originCountry + "/" + currancy + "/" + locale + "/" + origin + "/"  + destination + "/" + date
        response = requests.request("GET", myurl, headers=headers)
        results = json.loads(response.text)
        listlist.append(results)

        if("Quotes" in results):
            
            # Add to "airports" to create way to tie airport code back to airport name
            for Places in results["Places"]:
                airports[Places["PlaceId"]] = Places["Name"] 
            
            # Print data as it is pulled from API to track progress
            for Quotes in results["Quotes"]:
                print("***************")
                ori = Quotes["OutboundLeg"]["OriginId"]
                dest = Quotes["OutboundLeg"]["DestinationId"]
                price = Quotes["MinPrice"]
                print(f"${price} USD |{airports[ori]}| {airports[dest]} ")
                flights[airports[dest]] = price
                sleep(2.5)

print("-------------------------------")
print("DONE!")

        
    
```

```python
# Check list length and json format
dfdf = pd.json_normalize(listlist)
print(f"Total list length: {len(listlist)}")
listlist[2]
```

## Turn JSON into dataframe

```python
# Create list to cross-reference airport ID and IATA code
iata_list = {}

for flight in listlist:
    try:
        iata_list[flight['Places'][0]["PlaceId"]] = flight['Places'][0]["IataCode"]
        iata_list[flight['Places'][1]["PlaceId"]] = flight['Places'][1]["IataCode"]
    except:
        pass
    
print(len(iata_list))
```

```python
flight_list = []

for flight in listlist:
    flight_dict = {}
    try:
        
        # Add airport name and IATA code
        ori = flight['Quotes'][0]["OutboundLeg"]["OriginId"]
        dest = flight['Quotes'][0]["OutboundLeg"]["DestinationId"]
        flight_dict['origin_name'] = airports[ori]
        flight_dict['origin_iata'] = iata_list[ori]
        flight_dict['dest_name'] =  airports[dest]
        flight_dict['dest_iata'] = iata_list[dest]
        
        # Add price and carrier
        flight_dict['min_price_usd'] = flight['Quotes'][0]['MinPrice']
        flight_dict['carrier'] = flight['Carriers'][0]['Name']
        
        flight_list.append(flight_dict)
    
    except:
        pass
    
flightflight = pd.DataFrame(flight_list)
print(f" Length: {len(flight_list)} entries")
flightflight
```

```python
# Check to make sure data came through properly
origin_group = flightflight.groupby('origin_name').nunique()
origin_group

flightflight['dest_name'].nunique()
```

## Import airport GPS coordinates

```python
# Inport and clean GPS data
where_my_airport = pd.read_csv("https://datahub.io/core/airport-codes/r/airport-codes.csv")
where_my_airport = where_my_airport.drop(['municipality','ident','type','elevation_ft','continent','iso_country','iso_region','name','gps_code','local_code'], axis = 1)
where_my_airport = where_my_airport.dropna()
where_my_airport
```

## Merge flight data with GPS coordinates

```python
# Merge origin GPS
flight_df = pd.merge(flightflight,where_my_airport,how="inner",left_on="origin_iata",right_on="iata_code")
flight_df = flight_df.rename({'coordinates': 'origin_gps'}, axis='columns')

# Merge destination gps
flight_df = pd.merge(flight_df,where_my_airport,how="inner",left_on="dest_iata",right_on="iata_code")
flight_df = flight_df.rename({'coordinates': 'dest_gps'}, axis='columns')

# Clean
flight_df = flight_df.drop(['iata_code_y','iata_code_x'], axis = 1)

flight_df
```
